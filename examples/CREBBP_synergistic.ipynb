# CREBBP — dual-mechanism synergy + rescue

Two adjacent variants in CREBBP exon 2:

- `mut1` = `CREBBP:16:3758048:C:-`   (1bp deletion at the canonical acceptor)
- `mut2` = `CREBBP:16:3758049:C:T`   (SNV at the canonical acceptor)

This pair sits exactly at an annotated acceptor. We expect a mix of mechanisms across multiple sites:

1. The **canonical acceptor** at chr16:3,758,048 should be destroyed by both mutations.
2. A **cryptic acceptor** may become activated by one or both singles, then potentially rescued or further activated in the joint.

The classifier should surface the dominant synergistic event while preserving the rescue mechanism in the per-site table.

```python
from oncosplice import OncospliceEngine

eng = OncospliceEngine(splicing_engine='spliceai_pytorch')
pair = eng.analyze_pair(
    'CREBBP:16:3758048:C:-',
    'CREBBP:16:3758049:C:T',
    protein=False,
)
print('pair_classification:', pair.pair_classification)
```

## Two mechanisms at once

Look at every site the classifier flagged:

```python
sr = pair.site_residuals
epi = sr[sr.classification != 'non-epistatic']
epi[['position','site_type','annotated','ref','mut1','mut2','event','classification']].sort_values('position')
```

**Reading the table**: two distinct mechanisms at neighbouring cryptic acceptors.

- **A\* 3,758,027 — synergistic**. WT = 0.02, neither single produces > 0.10, but the joint event activates this cryptic acceptor to 0.99. This is real emergent synergy at a previously-silent site.
- **A\* 3,758,047 — rescue**. Each single alone activates this site to ~0.99 (creates a strong cryptic acceptor), but the joint event kills it (~0.0). The joint *rescues* the canonical splicing pattern at this position despite both singles wanting to disrupt it.

The pair-level call is **synergistic** because synergy outranks rescue in the priority hierarchy, but the dual-mechanism story is intact in the per-site table.
